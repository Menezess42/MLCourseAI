# Exploratory data analysis with Pandas

The core data structures in pandas are **Sereies** and **DataFrames**.

- Series:
    - Is a one-dimensional indexed array of a single data type
- DataFrame:
    - Is a two-dimensional tale whre each column contains data of the same type.
---
- Thinks of a **DataFrame** as a collection of **Series** objects.
---
- DataFrame are ideal for representing real-world data, with each row representing an instance (such as an oservation) and each column representing a feature of that instance.

### Grouping

- In general, grouping data in Pandas works as follows:
- **df.groupby(by=grouping_columns)[columns_to_show].function()**
- First the `groupby` method divides the grouping_columns by their values. They become a new index in the resulting dataframe.
- Then, the columns of interest area selected (`column_to_show`). If `columns_to_show` is not included, all non groupby clauses will be included.
- Finally, one or several functions are applied to the obtained groups per selected columns.

- Example:
- Group the data according to the values of the `churn` variable and display statistics of three columns in each group:

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
```

Let's do the same thing, but using `agg()`(Agregation)

```python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]
df.groupby(["Churn"])[columns_to_show].agg(["mean", "std", "min", "max"])
```

### Summary tables

- Suppose we want to see how the observations in our dataset are distributed in the context of two variables `Churn` and `International plan`.
- To do so, we can build a **contingency table** using the `crosstab` method:

```python
pd.crosstab(df["Churn"], df["International plan"])
```

```python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
```

We can see that most of the users are loyal and do not use additional services

- pivot table:
- The `pivot_table` method takes the following parameters:
- `values` -- a list of variables to calculate statistics for,
- `index` -- a list of variables to group data by,
- `aggfunc` -- what statistics we need to calculate for groups, e.g sum, mean, maximum, minimum, etc.

```python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
```

### DataFrame transformations

- Creating new dataFrame columns

```python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)

df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
```

- We can create column more easily without creating an intermediate

```python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
```

- To delete columns or rows, we can use the `drop` method
- Methods:
    - `axis`:
        - `1` to delete columns
        - `0` or `nothing` to delete rows
    - `inplace`:
        - Tells whether to change the original DataFrame
        - `inplace=False` doesn't change the exisiting DataFrame and returns a new one with the drop
        - `inplace=True` change the original DataFrame

```python
# Get rid of just created columns
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
# and here's how you can delete rows
df.drop([1, 2]).head()
```
